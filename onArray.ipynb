```python
import matplotlib.pyplot as plt
```

```python
# a) Line Plot
# Used to visualize trends over time or continuous data.


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y,color='green',marker='o')
#  color='green'
# marker='o',
# linestyle='--'
plt.title("Line Plot grap")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Scatter Plot
# Used to show the relationship between two variables.

x = [5, 7, 8, 10, 15]
y = [7, 14, 18, 21, 29]

plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
# Bar Chart
# Used for comparing different categories.

x = ['Apple', 'Banana', 'Carrot', 'Dragon_Fruit']
y = [23, 45, 56, 78]

plt.bar(x, y)
# color='orange'
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid()
plt.show()
# plt.savefig("barchart.png")
plt.savefig('file2.png')
```

```python
# Pie Chart
# Used to show percentage or proportional data.

labels = ['Apples', 'Bananas', 'Grapes', 'Oranges']
sizes = [30, 25, 25, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

```python
# Histogram
# Used to visualize the distribution of data.

import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=3, color='green',edgecolor='red')
# edgecolor
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

```

```python
# Box Plot
# Used to show distribution and detect outliers.
data = [20, 25, 30, 22, 35, 50, 55, 60, 100]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()


```

```python
# Subplots
# Display multiple plots in a single figure.

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 9]

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red')
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
# Styling the Plots

plt.style.use('ggplot')
plt.plot(x, y1, label='Prime Numbers')
plt.legend()
plt.grid(True)
plt.show()

```
