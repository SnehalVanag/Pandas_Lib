```python
import pandas as pd

data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance', 'IT'],
    'Employee': ['Snehal', 'Ishwari', 'Sayali', 'Komal', 'Rohan', 'Meena', 'Amit'],
    'Salary': [60000, 55000, 50000, 52000, 65000, 70000, 58000],
    'Experience': [2, 3, 4, 2, 5, 6, 1]
}

df = pd.DataFrame(data)
print(df)
```

```python
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

```python
summary = df.groupby('Department').agg({
    'Salary': ['mean', 'max', 'min'],
    'Experience': 'sum'
})
print(summary)
```
