```python
# Introduction

# Built on Matplotlib
# Provides a high-level API for beautiful, statistical plots.
# Works directly with Pandas DataFrames
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# df = sns.load_dataset("tips")  # sample dataset
df = sns.load_dataset("tips")  # sample dataset
print(df.head())
```

```python
# Scatter Plot
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex')
plt.title("Total Bill vs Tip")
plt.grid()
plt.show()
plt.savefig('filename.png')

# add hue
```

```python
# Line plot

sns.lineplot(x='size', y='tip', data=df)
plt.title("Tip vs Size")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df, palette='hot')
plt.show()
```

```python
# count plot
sns.countplot(x='day', data=df, palette='pastel')
plt.title("Count of Days")
plt.show()
```

```python
# BOx plot
sns.boxplot(x='day', y='total_bill', data=df, palette='Set2')
plt.title("Box Plot of Bills per Day")
plt.show()
```

```python
# Heatmap
# Visualizes correlation between numeric features.

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap Example")
plt.show()
```

```python
# Pair Plot
# Shows pairwise relationships across the dataset.

sns.pairplot(df,hue='sex')
plt.show()
```
